```python
temp = []
for i in df['sex']:
    if i =='female':
        temp.append(1)
    elif i == 'male':
        temp.append(0)
    else:
        temp.append(np.nan)
df['sex'] = temp
df
```

```python

df.info()
```

```python
#혈액형의 A,B,O,AB는 1,2,3,4로 바꾸면 우선순위가 생겨서 문제가 생김. 리스트로 [1000],[0100]...으로 바꾸기

df['survived'] = df['survived'].astype('float')
df['pclass'] = df['pclass'].astype('float')
df['sex'] = df['sex'].astype('float')
df['sibsp'] = df['sibsp'].astype('float')
df['parch'] = df['parch'].astype('float')
df['fare'] = df['fare'].astype('float')
```

```python
df=df[df['age'].notnull()]
df=df[df['fare'].notnull()]
```

```python
#시각화
#warnings 메세지 출력 안하기
import warnings
warnings.filterwarnings('ignore')
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips)
```

```python
g = sns.FacetGrid(tips, col='time', aspect=.7)
g.map(sns.boxplot, 'sex', 'total_bill', 'smoker').despine(left=True).add_legend(title='smoker')
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips, color='.5')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day',y='total_bill', data=tips, color='.25')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```
